### Define all variables here for Supplementary Figure 7b (Endogenous CKB)


```python
## Define all variables here for Supplementary Figure 7b

target = 'CKB' # define target gene name, should be the same as in bowtie reference file
sBowtieFasta = '/usersdata/joelamkc/genome/251029_RSEfig2_index/251029_RSE_fig2.fa' # define the path to the bowtie reference file
nTarget2Pos = 693 # define the crRNA position (3'end of the crRNA binding position)
lSamFiles = glob.glob('/usersdata/joelamkc/project/Cas13/results/251105_novaseq83_CKB_analysis_v2/data/*CKB.bt2output.simple.txt') # define the path to the aligment text files
sMatchFilter = '50M' # define the alignment score
sExcelOutputFile = '/usersdata/joelamkc/project/paper/cas13clvPos/cleancodes/output/figS7b.xlsx' # define the path to the destination in which you want to save the excel files

## to get the target mRNA sequence

for record in SeqIO.parse(open(sBowtieFasta, 'r'), 'fasta'):
    sName = str(record.id)
    sSeq = str(record.seq)
    if sName == target:
        sRefSeq = sSeq

dTargetGene = {target:target}
```

## Start the analysis here

```python
## Analyze the cleavage preference and output results in the form of spreadsheets

lSamDf = []
for sSamFile in lSamFiles:
    print(sSamFile)
    
    sFileName = sSamFile.split('/')[-1].split('.bt2output')[0]        
    sName = sFileName.split('.')[0]
    sTargetGeneName = dTargetGene[target]
    
    df = pd.read_table(sSamFile, header=None, sep=' ', names=['ReadNameCollapsed', 'Gene', 'Pos', 'MatchInfo', 'Seq'])                
    df['SampleName'] = sName
    df['SortedReadOrder'] = df.apply(lambda row: str(row['ReadNameCollapsed'].split('-')[0]), axis=1)
    df['CollapsedReadCount'] = df.apply(lambda row: int(row['ReadNameCollapsed'].split('-')[1]), axis=1)

    df['TargetPos'] = df.apply(lambda row: nTarget2Pos, axis=1)
    
    print('--')
    print(sName)
    print('all row numbers: ', len(df))
    print('all row read counts: ', df['CollapsedReadCount'].sum())
    
    df = df[df['Gene'] == sTargetGeneName]
    print('row numbers after selecting target genes: ', len(df))
    print('row read counts after selecting target genes: ', df['CollapsedReadCount'].sum())

    df = df[df['MatchInfo'] == sMatchFilter]
    print('row numbers after selecting a match filter: ', len(df))
    print('row read counts after selecting a match filter: ', df['CollapsedReadCount'].sum())

    # Get the nearby sequence around the cleavage site
    
    df['aroundClvSite'] = df.apply(lambda row: sRefSeq[row['Pos']-5: row['Pos']+3], axis=1) # 4nt before the cleavage and 4nt after the cleavage
    df['3ntMotif'] = df.apply(lambda row: sRefSeq[row['Pos']-3: row['Pos']], axis=1) # 2nt before the cleavage and 1nt after the cleavage    
    df['distance'] = df.apply(lambda row: row['Pos'] - row['TargetPos'], axis=1) # distance between binding and cleavage    
    df['MotifDist'] = df.apply(lambda row: row['3ntMotif'] + ', ' + str(row['distance']) + 'nt', axis=1)
    
    lSamDf.append((sName, df))

lSamDfSorted = sorted(lSamDf)

# Write an Excel file

lSubset = ['SampleName', 'Gene', 'Seq', 'CollapsedReadCount', 'SortedReadOrder', 'TargetPos', 'distance', 'aroundClvSite', '3ntMotif']
 
with pd.ExcelWriter(sExcelOutputFile, engine='xlsxwriter') as writer:            

    for sName, dfSam in lSamDfSorted:           

        dfSam = dfSam[lSubset]

        dfSam['SortedReadOrder'] = dfSam['SortedReadOrder'].astype(int)

        dfSam = dfSam.sort_values(by='SortedReadOrder', ascending=True)

        dfSam.to_excel(writer, sheet_name=sName, index=False)

    
```

```python
## Analyze the cleavage positions and show the distributions


# Count how many samples you have
n_samples = len(lSamDfSorted)

# Create ONE big figure with multiple samples rows and 1 column
fig, axes = plt.subplots(n_samples, 1, figsize=(8, 1.5 * n_samples),
                         sharex=True)                                  

# If only one sample, axes is not a list  make it always iterable
if n_samples == 1:
    axes = [axes]

for ax, (sName, dfSam) in zip(axes, lSamDfSorted):
    counts_by_pos = (dfSam.groupby('Pos')['CollapsedReadCount']
                         .sum()
                         .astype(int))

    nTargetPos = int(dfSam['TargetPos'].iloc[0])

    pos_min = counts_by_pos.index.min()
    pos_max = counts_by_pos.index.max()
    all_positions = np.arange(pos_min, pos_max + 1)                  
    counts_filled = counts_by_pos.reindex(all_positions, fill_value=0).values

    ax.bar(all_positions, counts_filled,
           width=0.98,                  
           color='#ee4444',
           edgecolor='black',
           linewidth=0.2,
           align='center')         

    ax.set_title(f'{sName} | End of crRNA = {nTargetPos}', fontsize=10, pad=8)
    ax.set_ylabel('Reads', fontsize=9)
    ax.axvline(nTargetPos, color='black', linestyle='--', linewidth=1.4, alpha=0.9)

    ax.spines[['top', 'right']].set_visible(False)
    ax.tick_params(axis='both', which='major', labelsize=8)
    ax.set_ylim(0, None)
    ax.set_xlim(650, 950)

axes[-1].set_xlabel('Nucleotide position', fontsize=10)
for ax in axes:
    ax.set_xlim(650, 950)

plt.tight_layout(pad=1.0)
#plt.savefig(" ", dpi=600, bbox_inches='tight')
plt.show()
```
